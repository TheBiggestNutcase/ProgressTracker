```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def add_progress(date, study_hours, work_hours):
    try:
        # Read existing data
        df = pd.read_csv('progress.csv')
    except FileNotFoundError:
        # If the file doesn't exist, create a new DataFrame
        df = pd.DataFrame(columns=['date', 'study_hours', 'work_hours'])

    # Create a new DataFrame for the new progress
    new_entry = pd.DataFrame({'date': [date], 'study_hours': [study_hours], 'work_hours': [work_hours]})

    # Concatenate the new entry to the existing DataFrame
    df = pd.concat([df, new_entry], ignore_index=True)
    df.to_csv('progress.csv', index=False)
```

```python
# Function to plot progress
def plot_progress():
    try:
        df = pd.read_csv('progress.csv')
        df['date'] = pd.to_datetime(df['date'])  # Convert to datetime
        df.sort_values('date', inplace=True)  # Sort by date

        plt.figure(figsize=(10, 5))
        plt.plot(df['date'], df['study_hours'], marker='o', label='Study Hours')
        plt.plot(df['date'], df['work_hours'], marker='o', label='Work Hours')
        plt.title('Daily Study and Work Hours')
        plt.xlabel('Date')
        plt.ylabel('Hours')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid()
        plt.tight_layout()
        plt.show()
    except FileNotFoundError:
        print("No data to plot. Please add some progress first.")
```

```python
# Main loop to interact with the user
def main():
    while True:
        action = input("Enter 'add' to add progress, 'plot' to view progress, or 'quit' to exit: ")
        if action == 'add':
            date = input("Enter the date (YYYY-MM-DD): ")
            study_hours = float(input("Enter study hours: "))
            work_hours = float(input("Enter work hours: "))
            add_progress(date, study_hours, work_hours)
        elif action == 'plot':
            plot_progress()
        elif action == 'quit':
            break
        else:
            print("Invalid input. Please try again.")
```

```python
if __name__ == "__main__":
    main()
```

```python
# Function to add progress
def add_progress(date, study_hours, work_hours, sleep_hours):
    try:
        # Read existing data
        df = pd.read_csv('progress.csv')
    except FileNotFoundError:
        # If the file doesn't exist, create a new DataFrame
        df = pd.DataFrame(columns=['date', 'study_hours', 'work_hours', 'sleep_hours'])

    # Create a new DataFrame for the new progress
    new_entry = pd.DataFrame({
        'date': [date], 
        'study_hours': [study_hours], 
        'work_hours': [work_hours], 
        'sleep_hours': [sleep_hours]
    })

    # Concatenate the new entry to the existing DataFrame
    df = pd.concat([df, new_entry], ignore_index=True)
    df.to_csv('progress.csv', index=False)
```

```python
# Function to plot progress
def plot_progress():
    try:
        df = pd.read_csv('progress.csv')
        df['date'] = pd.to_datetime(df['date'])  # Convert to datetime
        df.sort_values('date', inplace=True)  # Sort by date

        # Create subplots
        fig, axs = plt.subplots(1, 3, figsize=(18, 5))

        # Plot study hours
        axs[0].plot(df['date'], df['study_hours'], marker='o', color='blue')
        axs[0].set_title('Daily Study Hours')
        axs[0].set_xlabel('Date')
        axs[0].set_ylabel('Hours')
        axs[0].grid()

        # Plot work hours
        axs[1].plot(df['date'], df['work_hours'], marker='o', color='green')
        axs[1].set_title('Daily Work Hours')
        axs[1].set_xlabel('Date')
        axs[1].set_ylabel('Hours')
        axs[1].grid()

        # Plot sleep hours
        axs[2].plot(df['date'], df['sleep_hours'], marker='o', color='purple')
        axs[2].set_title('Daily Sleep Hours')
        axs[2].set_xlabel('Date')
        axs[2].set_ylabel('Hours')
        axs[2].grid()

        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    except FileNotFoundError:
        print("No data to plot. Please add some progress first.")
```

```python
# Main loop to interact with the user
def main():
    while True:
        action = input("Enter 'add' to add progress, 'plot' to view progress, or 'quit' to exit: ")
        if action == 'add':
            date = input("Enter the date (YYYY-MM-DD): ")
            study_hours = float(input("Enter study hours: "))
            work_hours = float(input("Enter work hours: "))
            sleep_hours = float(input("Enter sleep hours: "))
            add_progress(date, study_hours, work_hours, sleep_hours)
        elif action == 'plot':
            plot_progress()
        elif action == 'quit':
            break
        else:
            print("Invalid input. Please try again.")
```

```python
if __name__ == "__main__":
    main()
```
